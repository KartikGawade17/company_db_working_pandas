```python
import pandas as pd
```

```python
customers = pd.read_csv("customers.csv")
departments = pd.read_csv("departments.csv")
employees = pd.read_csv("employees.csv")
orders = pd.read_csv("orders.csv")
products = pd.read_csv("products.csv")
sales = pd.read_csv("sales.csv")
```

```python
# Show the department name and average salary of employees in each department, sorted from highest average salary to lowest.
departments.merge(employees, on = "department_id").groupby("department_name")["salary"].mean().sort_values(ascending=False)
```

```python
# Show the product name, total quantity sold, and total revenue generated for each product.
product_order_merge = products.merge(orders, on="product_id")
```

```python
product_order_merge.head(3)
```

```python
#Q.5 Find the employee who has generated the highest total sales amount from the sales table.

sales.merge(employees, left_on = "employee_id", right_on = "employee_id").groupby("employee_name")["sale_amount"].sum().sort_values(ascending=False).head(1)
```

```python
#Q.6 Show the city-wise count of customers and sort it from highest to lowest.
customers["city"].value_counts(ascending=False)
```

```python
# Find the products whose stock quantity is less than 20 and price is greater than 1000.
products[(products["stock_quantity"] < 20) & (products["price"] > 1000)] [["product_name", "price", "stock_quantity"]]
```
